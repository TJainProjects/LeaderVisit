```python
from transformers import pipeline
classifier = pipeline("zero-shot-classification", model="facebook/bart-large-mnli")
```

```python
import torch
print("GPU available:", torch.cuda.is_available())
```

```python
labels = [
    "0: No data",
    "1: Symbolic",
    "2: Economic/diplomatic deal",
    "3: Strategic "
]

def classify_visit(text):
    if pd.isna(text) or text.strip() == '':
        return 0
    result = classifier(text, labels)
    print(result)
    return int(result['labels'][0].split(":")[0])
```

```python
df['Visit_Intensity'] = df['Details'].apply(classify_visit)
```
